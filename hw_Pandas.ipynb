#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
students_info = ['Students_info_0.csv', 'Students_info_1.csv', 'Students_info_2.csv', 
                 'Students_info_3.csv', 'Students_info_4.csv', 'Students_info_5.csv', 
                 'Students_info_6.csv', 'Students_info_7.csv', 'Students_info_8.csv',
                 'Students_info_9.csv']

students_marks = ['Students_marks_0.csv', 'Students_marks_1.csv', 'Students_marks_2.csv',
                  'Students_marks_3.csv', 'Students_marks_4.csv', 'Students_marks_5.csv',
                  'Students_marks_6.csv', 'Students_marks_7.csv', 'Students_marks_8.csv',
                  'Students_marks_9.csv']



students = list(zip(students_info, students_marks))

df = pd.DataFrame() # создаем пустой датафрейм, чтобы дальше в него "вписать" данные 

for i in students:
    df_i = pd.merge(pd.read_csv(i[0]), pd.read_csv(i[1]), on = 'index')
    df = pd.concat([df, df_i], ignore_index = True)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
start_time_3 = time.time()

pairs = list(zip(df['race/ethnicity'].values, df['reading score'].values)) 

slovar = {}

for group, score in pairs:
    if group in slovar:
        slovar[group].append(score)
    else:
        slovar[group] = [score]
        
slovar = dict(sorted(slovar.items()))

for key, val in slovar.items():
    print(f'{key}: {sum(val) / len(val)}')

print('\n')
print(f'Время выполнения: {time.time() - start_time_3}')
```

```python
start_time_4 = time.time()

pair = list(zip(df['education'].values, df['writing score'].values))

slovar = {}

for group, score in pair:
    if group in slovar:
        slovar[group].append(score)
    else:
        slovar[group] = [score]
        
slovar = dict(sorted(slovar.items()))

for key, val in slovar.items():
    print(f'{key}: {min(val)}')

print('\n')
print(f'Время выполнения: {time.time() - start_time_4}')
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    elif 80 <= average_mark < 90:
        return 'B'
    elif 70 <= average_mark < 80:
        return 'C'
    elif 60 <= average_mark < 70:
        return 'D'
    elif 50 <= average_mark < 60:
        return 'E'
    else:
        return 'F'

df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis = 1).apply(GetGrade)


df
```

```python
df.pivot_table(index = ['Grade'], aggfunc = {'Grade':'count'})
```
